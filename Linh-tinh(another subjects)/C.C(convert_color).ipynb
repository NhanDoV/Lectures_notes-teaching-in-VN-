```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

**Display the original image**
- In opencv; the order (when using `cv2.imshow`) of `image_channel` is `Blue-Green-Red`; but in matplotlib.pyplot; the order is `Red-Green-Blue`.

- Firstly, let's see the original image

```python
path = "D://ghost.jpg"
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

**`The image_shape`** = height, width, number_of_channel

```python
#path = "D://mask.png"
path = "D://ghost.jpg"
img = cv2.imread(path)
img.shape
```

**playing with the order of color-channel**

```python
height, width = img.shape[0:2]

img_1 = cv2.imread(path)
img_1[:,:,0] = img_1[:,:,2]

img_2 = cv2.imread(path)
img_2[:,:,1] = img_2[:,:,2]

img_3 = cv2.imread(path)
img_3[:,:,0] = img_3[:,:,2]
img_3[:,:,1] = img_3[:,:,2]

## Row 1
img = cv2.imread(path) # hnh gc
img_mer1 = np.stack([img, img_1, img_2, img_3], axis=1).reshape(height, width*4, 3)
plt.imshow(img_mer1)
```

```python
img = ~img
img_1 = ~img_1
img_2 = ~img_2 ##cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
img_3 = ~img_3 ## cv2.cvtColor(img_3, cv2.COLOR_BGR2HSV)

## Row 2
#img = cv2.imread(path); img_3 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_mer2 = np.stack([img, img_1, img_2, img_3], axis=1).reshape(height, width*4, 3)
plt.imshow(img_mer2)
```

```python
img_1 = cv2.imread(path)
img_1[:,:,0] = img_1[:,:,2]; im1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)

img_2 = cv2.imread(path)
img_2[:,:,1] = img_2[:,:,2]; im2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)

img_3 = cv2.imread(path)
img_3[:,:,0] = img_3[:,:,2]
img_3[:,:,1] = img_3[:,:,2]; im3 = cv2.cvtColor(img_3, cv2.COLOR_BGR2RGB)

## Row 4
img = cv2.imread(path); img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_mer4 = np.stack([img, im1, im2, im3], axis=1).reshape(height, width*4, 3)
plt.imshow(img_mer4)
```

```python
img_mer3 = ~img_mer4
plt.imshow(img_mer3)
```

`Merged all rows contains the differents_color_style of image`

```python
img_mer = np.stack([img_mer1, img_mer2, img_mer4, img_mer3], axis=0).reshape(height*4, width*4, 3)
plt.figure(figsize = (8, 8))
plt.imshow(img_mer)
```
